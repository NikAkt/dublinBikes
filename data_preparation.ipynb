Check for missing values:


```python
print(a.isna().sum())
print("=====================================")
print(w.isna().sum())
```

```python
# Convert 'curr_time' to datetime format
w["curr_time"] = pd.to_datetime(w["curr_time"])

# Floor the 'curr_time' to the nearest hour
w["curr_time"] = w["curr_time"].dt.floor("h")

a["curr_time"] = pd.to_datetime(a["curr_time"])

a["curr_time"] = a["curr_time"].dt.floor("h")
```

```python
merged_df = pd.merge(w, a, on="curr_time", how="inner")
```

Check for duplicate rows:


```python
# Find duplicate rows
# Find all instances of duplicate rows
duplicate_rows1 = merged_df[merged_df.duplicated(keep=False)]
duplicate_rows1
```

```python
sorted_df = duplicate_rows1.sort_values(by="last_update")
sorted_df
```
