```python
import pandas as pd
```

```python
w = pd.read_csv("weather.csv")
a = pd.read_csv("availability.csv")
```

Check the layout of the data right now:


```python
print(a.shape)
print(w.shape)
```

```python
a.tail()
```

```python
w.tail()
```

Check for missing values:


```python
print(a.isna().sum())
print("=====================================")
print(w.isna().sum())
```

```python
# Convert 'curr_time' to datetime format
w["curr_time"] = pd.to_datetime(w["curr_time"])

# Floor the 'curr_time' to the nearest hour
w["curr_time"] = w["curr_time"].dt.floor("h")

a["curr_time"] = pd.to_datetime(a["curr_time"])

a["curr_time"] = a["curr_time"].dt.floor("h")
```

```python
a_dupes = a[a.duplicated(keep=False)]
a_dupes
```

```python
w_dupes = w[w.duplicated(keep=False)]
w_dupes
```

```python
# Drop duplicates but keep the first occurrence
w_no_dupes = w.drop_duplicates(keep="first")
w_dupes.shape
```

```python
a_no_dupes = a.drop_duplicates(keep="first")
a_no_dupes.shape
```

```python
merged_df = pd.merge(w_no_dupes, a_no_dupes, on="curr_time", how="inner")
merged_df.shape
```

```python
merged_df.head()
```

```python
merged_df.tail()
```

Check for duplicate rows:


```python
duplicate_rows = merged_df[merged_df.duplicated(keep=False)]
duplicate_rows
```

```python
merged_df["lon"] = merged_df["lon"].astype("category")
merged_df["lat"] = merged_df["lat"].astype("category")
merged_df["id"] = merged_df["id"].astype("category")
merged_df["weather_description"] = merged_df["weather_description"].astype("category")
merged_df["number"] = merged_df["number"].astype("category")

merged_df["temp"] = merged_df["temp"].astype("float64")
merged_df["feels_like"] = merged_df["feels_like"].astype("float64")
merged_df["wind_speed"] = merged_df["wind_speed"].astype("float64")

merged_df["humidity"] = merged_df["humidity"].astype("int64")
merged_df["available_bikes"] = merged_df["available_bikes"].astype("int64")
merged_df["available_bike_stands"] = merged_df["available_bike_stands"].astype("int64")

merged_df["curr_time"] = pd.to_datetime(merged_df["curr_time"])
merged_df["last_update"] = pd.to_datetime(merged_df["last_update"])
```

```python
merged_df["date"] = merged_df["curr_time"].dt.date
merged_df["time"] = merged_df["curr_time"].dt.time
```
